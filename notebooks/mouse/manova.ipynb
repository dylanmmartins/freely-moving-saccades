```python
import numpy as np
from scipy.io import savemat
from statsmodels.formula.api import ols
import statsmodels.api as sm
import pandas as pd
```

```python
# sf = np.load(r'C:\Users\dmartins\Desktop\mouse_sf.npy', allow_pickle=True)
# tf = np.load(r'C:\Users\dmartins\Desktop\mouse_tf.npy', allow_pickle=True)
sf = np.load(r'C:\Users\dmartins\Desktop\marmo.npy', allow_pickle=True)
```

```python
df = pd.DataFrame()
for k in range(4):
    _df = pd.DataFrame(sf[k], columns=np.array([1,2,4,8,16]).astype(str))
    _df.insert(0, 'cluster', ['early','late','biphasic','negative'][k])
    df = pd.concat([df, _df])
df.reset_index(inplace=True)

df1 = pd.DataFrame()
i = 0
for ind, row in df.iterrows():
    for _tmpuse in np.array([1,2,4,8,16]).astype(str):
        df1.at[i,'cluster'] = row['cluster']
        df1.at[i,'index'] = row['index']
        df1.at[i,'response'] = row[_tmpuse]
        df1.at[i,'sf'] = _tmpuse
        i += 1
```

```python
df1[df1['sf']!='16']
```

```python
df1
```

```python
model = ols('response ~ C(sf) + C(cluster) + C(sf):C(cluster)', data=df1[df1['sf']!='16']).fit()
print(sm.stats.anova_lm(model, typ=2))
```

```python
tf = np.load(r'C:\Users\dmartins\Desktop\mouse_tf.npy', allow_pickle=True)
```

```python

df = pd.DataFrame()
for k in range(4):
    _df = pd.DataFrame(tf[k], columns=['low','high'])
    _df.insert(0, 'cluster', ['early','late','biphasic','negative'][k])
    df = pd.concat([df, _df])
df.reset_index(inplace=True)

df1 = pd.DataFrame()
i = 0
for ind, row in df.iterrows():
    for _i in range(2):
        _tmpuse = ['low','high'][_i]
        df1.at[i,'cluster'] = row['cluster']
        df1.at[i,'index'] = row['index']
        df1.at[i,'response'] = row[_tmpuse]
        df1.at[i,'sf'] = _tmpuse
        i += 1
```

```python
model = ols('response ~ C(sf) + C(cluster) + C(sf):C(cluster)', data=df1).fit()
print(sm.stats.anova_lm(model, typ=2))
```
