```python
hffm = pd.read_pickle('/home/niell_lab/Data/freely_moving_ephys/batch_files/062022/hffm_062022_gt.pickle')
```

```python
hffm['session'].unique()
```

```python
unit = hffm[hffm['session']=='070921_J553RT_control_Rig2'].iloc[0].copy()
```

```python
t0 = unit['t0']
```

```python
gazeL = unit['FmLt_gazeshift_left_saccTimes_dHead1'].copy().astype(float)
gazeR = unit['FmLt_gazeshift_right_saccTimes_dHead1'].copy().astype(float)
compL = unit['FmLt_comp_left_saccTimes_dHead1'].copy().astype(float)
compR = unit['FmLt_comp_right_saccTimes_dHead1'].copy().astype(float)
```

```python
stimFr = {}
labels = ['gazeL','gazeR','compL','compR']
for i, times in enumerate([gazeL, gazeR, compL, compR]):
    label = labels[i]
    frames = [int(np.nanargmin(np.abs(worldT_-t))-startInd) for t in times if (t<endT) and (t>startT)]
    stimFr[label] = frames
```

```python
with open('/home/niell_lab/Desktop/worldcam_playback_saccframes.json', 'w') as f:
    json.dump(stimFr, f)
```
