```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = np.load('/home/niell_lab/Mounts/Goeppert/nlab-nas/freely_moving_ephys/ephys_recordings/041422/J599RT/addtlhf_props2.npz')
```

```python
data['rc'].shape
```

```python
all_psth = data['rc'].copy()
```

```python
bins = np.arange(-1000,1001,1)

fig, axs = plt.subplots(10,8, dpi=300, figsize=(15,15))

for i  in range(80):
    psth = all_psth[i,:]
    
    row = int(np.floor(i/8))
    col = int(i%8)
    
    axs[row,col].plot(bins, psth)
    axs[row,col].vlines(0, 0, np.max(psth)*1.1, color='k', linestyle='dashed')
    axs[row,col].set_ylim([0, np.max(psth)*1.1])
    axs[row,col].set_xticklabels([])
    axs[row,col].set_xlim([-500,500])

fig.tight_layout()
```
